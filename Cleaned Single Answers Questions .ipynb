```python
df = df[~df["questionnaire_login_year_group_name"].str.contains("SEND PE", na=False)]
```

```python
df = df[df["school_classification_list"].str.lower().isin(["primary", "secondary"])]
```

```python
df = df.rename(columns={"Please select your gender": "Gender",
                        "questionnaire_login_year_group_name": "Year_group",
                        "school_classification_list": "School classifications"
                       })
```

```python
df = df[df["Gender"].str.lower().isin(["male", "female"])]
```

```python
df = df.replace("â€™", "’", regex=True)
```

```python
df["In a normal week, how often do you have breakfast (this means not just a snack or a drink)?"] = df["In a normal week, how often do you have breakfast (this means not just a snack or a drink)?"].str.extract(r"\((.*?)\)", expand=False)
```

```python
df["How many minutes a day, do you usually spend sitting down watching TV/playing computer games/using the internet/using a mobile phone?"] = df["How many minutes a day, do you usually spend sitting down watching TV/playing computer games/using the internet/using a mobile phone?"].str.extract(r"\((.*?)\)", expand=False)
```

```python
df["How many minutes a day, do you usually spend sitting down watching TV/playing computer games/using the internet/using a mobile phone?"] = df["How many minutes a day, do you usually spend sitting down watching TV/playing computer games/using the internet/using a mobile phone?"].str.replace("1 minute", "", regex=False).str.strip()
```

```python
df["In a normal week how many times are you physically active for 30 minutes or more?"] = df["In a normal week how many times are you physically active for 30 minutes or more?"].replace("0 (Never)", "0")
```
